```python
import librosa
import torch
```

```python
import fnmatch
import os

files = []
for root, dirnames, filenames in os.walk("MP3-Example"):
    for filename in fnmatch.filter(filenames, "*mp3"):
        files.append(os.path.join(root, filename))
```

```python
import tqdm

train_arr = list()
for files in tqdm.tqdm(files):
    audio, sample_rate = librosa.load(files)
    audio = audio[:600000].squeeze()
    mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate)
    train_arr.append(torch.tensor(mfcc))
```

```python
train_arr
```

```python
mfcc_tensor = torch.stack(train_arr)
```
