```python
genres_list = ['classical', 'ambient', 'electronic', 'folk', 'rap', 'indie', 'pop', 'instrumental']

def extract_genre(tag): #функция для заполнения жанров по тегам
    if isinstance(tag, str):
        if 'rock' in tag:
            return 'rock'
        elif 'metal' in tag:
            return 'metal'
        for genre in genres_list:
            if genre in tag:
                return genre
    return ''
    
data['genre'] = data['tags'].apply(extract_genre)
```

```python
data['genre'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier #заполнение пропусков

train = data[data['genre'] != '']
test = data[data['genre'] == '']

X_train = train[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
X_test = test[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
y_train = train['genre']

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)

model = RandomForestClassifier()
model.fit(X_train, y_train)

data.loc[data['genre'] == '', 'genre'] = model.predict(X_test)
```

```python
data['genre'].value_counts()
```

```python
from sklearn.metrics import accuracy_score, classification_report #проверка метрик на сущетсвующих до этого значениях жанра

train_sample = train.sample(frac=0.8, random_state = 42)
test_sample = train.drop(train_sample.index)

X_train = train_sample[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
X_test = test_sample[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]

X_train = pd.get_dummies(X_train)
y_train = train_sample['genre']
X_test = pd.get_dummies(X_test)
y_test = test_sample['genre']

X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print('accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
def classify_emotions(energy, valence): #функция для определения эмоции трека
    if energy > 0.5 and valence > 0.5:
        return 'joy'
    elif energy < 0.5 and valence > 0.5:
        return 'pleasure'
    elif energy < 0.5 and valence < 0.5:
        return 'sad'
    else:
        return 'anger'
```

```python
data['emotion'] = data.apply(lambda x: classify_emotions(x.energy, x.valence), axis = 1)
```

Создаем фейковый timestamp для последовательности взаимодействия юзера с песней, это нужно для SasRec

```python
df['timestamp'] = df.groupby('user_id').cumcount() + 1
```

```python
df
```

```python
track_map = {track_id: idx for idx, track_id in enumerate(df['track_id'].unique())}
df['track_idx'] = df['track_id'].map(track_map)
```

```python
user_map = {user_id: idx for idx, user_id in enumerate(df['user_id'].unique())}
df['user_idx'] = df['user_id'].map(user_map)
```

```python
df_sorted = df.sort_values(by=['user_idx', 'timestamp'])
```

```python
df_sorted
```

```python
user_sequences = df_sorted.groupby('user_idx')['track_idx'].apply(list)
```

Получаем последовательность взаимодействия каждого юзера с треками в "хронологическом порядке"

```python
user_sequences
```

```python
result = []

for idx, sequence in user_sequences.items():
    result.extend([[idx, value] for value in sequence])

result_df = pd.DataFrame(result, columns=["Index", "Value"])
result_df = result_df.reset_index(drop=True)

with open("result.txt", "w") as f:
    f.write(result_df.to_string(index=False))
```
