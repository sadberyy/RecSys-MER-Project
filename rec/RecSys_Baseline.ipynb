```python
genres_list = [
    "classical",
    "ambient",
    "electronic",
    "folk",
    "rap",
    "indie",
    "pop",
    "instrumental",
]


def extract_genre(tag):  # функция для заполнения жанров по тегам
    if isinstance(tag, str):
        if "rock" in tag:
            return "rock"
        elif "metal" in tag:
            return "metal"
        for genre in genres_list:
            if genre in tag:
                return genre
    return ""


data["genre"] = data["tags"].apply(extract_genre)
```

```python
data["genre"].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier  # заполнение пропусков

train = data[data["genre"] != ""]
test = data[data["genre"] == ""]

X_train = train[
    [
        "danceability",
        "energy",
        "loudness",
        "speechiness",
        "acousticness",
        "instrumentalness",
        "liveness",
        "valence",
        "tempo",
    ]
]
X_test = test[
    [
        "danceability",
        "energy",
        "loudness",
        "speechiness",
        "acousticness",
        "instrumentalness",
        "liveness",
        "valence",
        "tempo",
    ]
]
y_train = train["genre"]

X_train = pd.get_dummies(X_train)


X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

model = RandomForestClassifier()
model.fit(X_train, y_train)

data.loc[data["genre"] == "", "genre"] = model.predict(X_test)
```

```python
def read_user_sequences_from_file(filepath):
    user_sequences = {}

    with open(filepath, "r") as file:
        next(file)
        for line in file:
            if not line:
                continue
            user_id, song_id = line.strip().split()  # Разделяем строку по пробелу
            user_id, song_id = int(user_id), int(song_id)  # Преобразуем в числа

            if user_id not in user_sequences:
                user_sequences[user_id] = (
                    []
                )  # Инициализируем пустой список для нового пользователя

            user_sequences[user_id].append(song_id)  # Добавляем песню к пользователю

    return user_sequences


filepath = "result.txt"
user_sequences = read_user_sequences_from_file(filepath)
# user_sequences
```

```python
import torch
import torch.nn as nn
import numpy as np
```

```python
class SASRec(nn.Module):
    def __init__(self, num_items, embedding_dim, num_heads, num_layers, max_seq_len):
        super(SASRec, self).__init__()
        self.item_embedding = nn.Embedding(num_items, embedding_dim)
        self.position_embedding = nn.Embedding(max_seq_len, embedding_dim)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim,
            nhead=num_heads,
            # num_encoder_layers = num_layers,
            # num_decoder_layers = 0,
            dim_feedforward=embedding_dim * 4,
            dropout=0.1,
            batch_first=True,
        )
        # self.transformer_block = nn.Transformer(
        #     d_model = embedding_dim,
        #     nhead = num_heads,
        #     num_encoder_layers = num_layers,
        #     num_decoder_layers = 0,
        #     dim_feedforward = embedding_dim * 4,
        #     dropout = 0.1,
        #     batch_first = True
        # )
        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layer, num_layers=num_layers
        )
        self.fc = nn.Linear(embedding_dim, num_items)

    def forward(self, seq):
        seq_len = seq.size(1)
        positions = torch.arange(seq_len, device=seq.device).unsqueeze(0).expand_as(seq)
        item_emb = self.item_embedding(seq)
        pos_emb = self.position_embedding(positions)
        x = item_emb + pos_emb  # сумм эмб позиций и эл-тов
        # x = self.transformer_block(x)
        x = self.transformer_encoder(x)
        x = self.fc(x)
        return x
```

ф-я для преобраз-я данных в посл-ти фикс длины

```python
def prepare_data(user_sequences, max_seq_len):
    information = []
    for user, sequence in user_sequences.items():
        for i in range(1, len(sequence)):
            target = sequence[i]
            # обрезка посл-ти
            context = sequence[max(0, i - max_seq_len) : i]
            # дополн до фикс длины
            padded_context = [0] * (max_seq_len - len(context)) + context
            information.append((padded_context, target))
    return information
```

ф-я для обучения модели

```python
import tqdm
```

```python
def train(model, data_loader, optimizer, criterion, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        for batch in tqdm.tqdm(data_loader):
            context, target = batch
            context = context.to(device)
            target = target.to(device)

            optimizer.zero_grad()
            output = model(context)
            loss = criterion(output[:, -1, :], target)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print("epoch", epoch + 1, "Loss ", total_loss)
```

Функция для предсказания:

```python
def predict(model, user_sequence, max_seq_len):
    model.eval()
    sequence = torch.tensor(user_sequence[-max_seq_len:]).unsqueeze(0)
    with torch.no_grad():
        scores = model(sequence)
        return torch.topk(scores[:, -1, :], k=10).indices.squeeze().tolist()
```

```python
import os

num_workers = os.cpu_count()


class SequenceDataset(Dataset):
    def __init__(self, information):
        self.information = information

    def __len__(self):
        return len(self.information)

    def __getitem__(self, idx):
        context, target = self.information[idx]
        return torch.tensor(context), torch.tensor(target)


max_seq_len = 10
prepared_data = prepare_data(user_sequences, max_seq_len)
dataset = SequenceDataset(prepared_data)
data_loader = DataLoader(
    dataset, batch_size=1024, shuffle=True, num_workers=num_workers
)
```

```python
def generate_recommendations_for_all_users(user_sequences, model, max_seq_len, device):
    recommendations_dict = {}
    for user_id in range(len(user_sequences)):
        user_sequence = user_sequences[user_id]
        user_sequence = torch.tensor(user_sequence).to(device)
        recommendations = predict(model, user_sequence, max_seq_len)
        recommendations_dict[user_id] = recommendations
    return recommendations_dict


relevant_items_dict = generate_recommendations_for_all_users(
    user_sequences, model, max_seq_len, device
)
print("Рекомендации для первого пользователя:", relevant_items_dict[0])
```

```python
actual_dict = data_txt.groupby("Index")["Value"].apply(list).to_dict()
# actual_dict
```

```python
def precision_at_k(recommended, actual, k=10):
    relevant_items = set(actual[:k])
    recommended_items = set(recommended[:k])

    return len(relevant_items & recommended_items) / k


def dcg_at_k(recommended, actual, k=10):
    dcg = 0
    for i, item in enumerate(recommended[:k]):
        if item in actual:
            dcg += 1 / np.log2(i + 2)
    return dcg


def ndcg_at_k(recommended, actual, k=10):
    dcg = dcg_at_k(recommended, actual, k)
    ideal_recommendations = sorted(actual, reverse=True)
    idcg = dcg_at_k(recommended, ideal_recommendations, k)
    return dcg / idcg if idcg > 0 else 0


def map_at_k(predicted_dict, actual_dict, k=10):
    ap_sum = 0
    num_users = len(predicted_dict)
    for user_id in predicted_dict:
        recommended = predicted_dict[user_id]
        actual = actual_dict.get(user_id, [])
        ap_sum += precision_at_k(recommended, actual, k)
    return ap_sum / num_users
```

```python
map_10 = map_at_k(relevant_items_dict, actual_dict, k=10)
print(f"MAP@10: {map_10}")
```

rel — релевантность элемента на позиции
(обычно это бинарное значение 1 для релевантного элемента и 0 для нерелевантного или оценки релевантности).
k — количество позиций в списке, на которые мы смотрим (k=10 для NDCG@10).
log2(i+1) — коэффициент, уменьшающий вклад элемента, чем дальше он находится в списке. В IDCG rel - релевантности отсортированы по убыванию (идеальный случай, когда все релевантные элементы находятся в начале списка)
