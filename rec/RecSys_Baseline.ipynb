```python
genres_list = [
    "classical",
    "ambient",
    "electronic",
    "folk",
    "rap",
    "indie",
    "pop",
    "instrumental",
]


def extract_genre(tag):  #      
    if isinstance(tag, str):
        if "rock" in tag:
            return "rock"
        elif "metal" in tag:
            return "metal"
        for genre in genres_list:
            if genre in tag:
                return genre
    return ""


data["genre"] = data["tags"].apply(extract_genre)
```

```python
class SASRec(nn.Module):
    def __init__(self, num_items, embedding_dim, num_heads, num_layers, max_seq_len):
        super(SASRec, self).__init__()
        self.item_embedding = nn.Embedding(num_items, embedding_dim)
        self.position_embedding = nn.Embedding(max_seq_len, embedding_dim)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim,
            nhead=num_heads,
            # num_encoder_layers = num_layers,
            # num_decoder_layers = 0,
            dim_feedforward=embedding_dim * 4,
            dropout=0.1,
            batch_first=True,
        )
        # self.transformer_block = nn.Transformer(
        #     d_model = embedding_dim,
        #     nhead = num_heads,
        #     num_encoder_layers = num_layers,
        #     num_decoder_layers = 0,
        #     dim_feedforward = embedding_dim * 4,
        #     dropout = 0.1,
        #     batch_first = True
        # )
        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layer, num_layers=num_layers
        )
        self.fc = nn.Linear(embedding_dim, num_items)

    def forward(self, seq):
        seq_len = seq.size(1)
        positions = torch.arange(seq_len, device=seq.device).unsqueeze(0).expand_as(seq)
        item_emb = self.item_embedding(seq)
        pos_emb = self.position_embedding(positions)
        x = item_emb + pos_emb  #     -
        # x = self.transformer_block(x)
        x = self.transformer_encoder(x)
        x = self.fc(x)
        return x
```

```python
def prepare_data(user_sequences, max_seq_len):
    information = []
    for user, sequence in user_sequences.items():
        for i in range(1, len(sequence)):
            target = sequence[i]
            #  -
            context = sequence[max(0, i - max_seq_len) : i]
            #    
            padded_context = [0] * (max_seq_len - len(context)) + context
            information.append((padded_context, target))
    return information
```

```python
from tqdm import tqdm
```

```python
def train(model, data_loader, optimizer, criterion, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        for batch in tqdm(data_loader):
            context, target = batch
            context = context.to(device)
            target = target.to(device)

            optimizer.zero_grad()
            output = model(context)
            loss = criterion(output[:, -1, :], target)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print("epoch", epoch + 1, "Loss ", total_loss)
```

```python
def predict(model, user_sequence, max_seq_len):
    model.eval()
    sequence = torch.tensor(user_sequence[-max_seq_len:]).unsqueeze(0)
    with torch.no_grad():
        scores = model(sequence)
        return torch.topk(scores[:, -1, :], k=10).indices.squeeze().tolist()
```

```python
def precision_at_k(recommended, actual, k=10):
    relevant_items = set(actual[:k])
    recommended_items = set(recommended[:k])

    return len(relevant_items & recommended_items) / k


def dcg_at_k(recommended, actual, k=10):
    dcg = 0
    for i, item in enumerate(recommended[:k]):
        if item in actual:
            dcg += 1 / np.log2(i + 2)
    return dcg


def ndcg_at_k(recommended, actual, k=10):
    dcg = dcg_at_k(recommended, actual, k)
    ideal_recommendations = sorted(actual, reverse=True)
    idcg = dcg_at_k(recommended, ideal_recommendations, k)
    return dcg / idcg if idcg > 0 else 0


def map_at_k(predicted_dict, actual_dict, k=10):
    ap_sum = 0
    num_users = len(predicted_dict)
    for user_id in predicted_dict:
        recommended = predicted_dict[user_id]
        actual = actual_dict.get(user_id, [])
        ap_sum += precision_at_k(recommended, actual, k)
    return ap_sum / num_users
```

```python
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader

MAX_LEN = 10
BATCH_SIZE = 1024

user_interactions = user_sequences

all_items = set(item for seq in user_interactions.values() for item in seq)
item2idx = {item: idx + 1 for idx, item in enumerate(sorted(all_items))}
idx2item = {idx: item for item, idx in item2idx.items()}
vocab_size = len(item2idx) + 1

user_sequences = [
    [item2idx[item] for item in seq] for seq in user_interactions.values()
]
```

```python
class BERT4RecDataset(Dataset):
    def __init__(self, sequences, max_len):
        self.sequences = sequences
        self.max_len = max_len

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        seq = self.sequences[idx][-self.max_len :]
        pad_len = self.max_len - len(seq)

        mask_idx = np.random.randint(len(seq))
        masked_seq = seq.copy()
        masked_seq[mask_idx] = 0

        masked_seq += [0] * pad_len
        seq += [0] * pad_len

        return torch.tensor(masked_seq, dtype=torch.long), torch.tensor(
            seq[mask_idx], dtype=torch.long
        )


dataset = BERT4RecDataset(user_sequences, MAX_LEN)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
from transformers import BertConfig, BertModel


class BERT4Rec(torch.nn.Module):
    def __init__(
        self, vocab_size, hidden_size=128, num_layers=1, num_heads=1, max_len=MAX_LEN
    ):
        super(BERT4Rec, self).__init__()
        config = BertConfig(
            vocab_size=vocab_size,
            hidden_size=hidden_size,
            num_hidden_layers=num_layers,
            num_attention_heads=num_heads,
            max_position_embeddings=max_len,
        )
        self.bert = BertModel(config)
        self.output_layer = torch.nn.Linear(hidden_size, vocab_size)

    def forward(self, input_ids):
        output = self.bert(input_ids).last_hidden_state
        return self.output_layer(output)


model = BERT4Rec(vocab_size)
```

```python
import torch.optim as optim
from sklearn.metrics import accuracy_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


def train_model(model, dataloader, optimizer, loss_fn, num_epochs):

    model.to(device)
    model.train()

    for epoch in range(num_epochs):
        running_loss = 0.0
        all_preds = []
        all_targets = []

        progress_bar = tqdm(dataloader, desc=f"Epoch {epoch + 1}/{num_epochs}")

        for masked_seq, target in progress_bar:
            masked_seq, target = masked_seq.to(device), target.to(device)

            optimizer.zero_grad()

            output = model(masked_seq)[:, -1, :]

            loss = loss_fn(output, target)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            _, predicted = torch.max(output, dim=1)
            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(target.cpu().numpy())

            progress_bar.set_postfix(loss=loss.item())

        epoch_loss = running_loss / len(dataloader)
        epoch_accuracy = accuracy_score(all_targets, all_preds)

        print(
            f"Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.4f}"
        )


optimizer = optim.Adam(model.parameters(), lr=1e-4)
loss_fn = torch.nn.CrossEntropyLoss()

train_model(model, dataloader, optimizer, loss_fn, num_epochs=2)
```

```python
def get_top_n_recommendations(model, user_seq, n):
    model.eval()
    device = next(model.parameters()).device

    input_seq = torch.tensor(user_seq).unsqueeze(0).to(device)
    output = model(input_seq)[:, -1, :]
    _, top_n_idx = torch.topk(output, n)

    return top_n_idx.squeeze().cpu().numpy()


user_seq = [1, 2, 1599, 3609, 3690, 3890]
top_recommendations = get_top_n_recommendations(model, user_seq, n=10)
print(top_recommendations)
```

```python
import torch
import torch.nn.functional as F


def pad_sequence(seq, max_len):
    return F.pad(torch.tensor(seq), (0, max_len - len(seq)), "constant", 0)


def get_top_n_recommendations(model, user_seq, n, max_len):
    model.eval()
    device = next(model.parameters()).device

    if len(user_seq) < max_len:
        user_seq = pad_sequence(user_seq, max_len)
    elif len(user_seq) > max_len:
        user_seq = user_seq[:max_len]

    input_seq = torch.tensor(user_seq).unsqueeze(0).to(device)
    output = model(input_seq)[:, -1, :]
    _, top_n_idx = torch.topk(output, n)

    return top_n_idx.squeeze().cpu().numpy()


max_len = 10

total_sequences = len(user_sequences)

all_recommendations = [
    get_top_n_recommendations(model, user_seq, n=10, max_len=max_len)
    for user_seq in tqdm(
        user_sequences, desc=" ", total=total_sequences
    )
]

print(all_recommendations[:5])
```

```python
import numpy as np


def apk(actual, predicted, k=10):
    if not actual:
        return 0.0

    predicted = predicted[:k]
    score = 0.0
    num_hits = 0.0

    for i, p in enumerate(predicted, start=1):
        if p in actual and p not in predicted[: i - 1]:
            score += num_hits / i

    return score / min(len(actual), k)


def mapk(actual_list, predicted_list, k=10):
    return np.mean([apk(a, p, k) for a, p in zip(actual_list, predicted_list)])


def dcg_at_k_another(relevance, k=10):
    relevance = np.array(relevance)[:k]
    return np.sum(relevance / np.log2(np.arange(2, len(relevance) + 2)))


def ndcg_at_k(actual, predicted, k=10):
    relevance = [1 if item in actual else 0 for item in predicted[:k]]
    ideal_relevance = sorted(relevance, reverse=True)

    dcg = dcg_at_k_another(relevance, k)
    idcg = dcg_at_k_another(ideal_relevance, k)

    return dcg / idcg if idcg > 0 else 0.0


def mean_ndcg(actual_list, predicted_list, k=10):
    return np.mean([ndcg_at_k(a, p, k) for a, p in zip(actual_list, predicted_list)])
```

```python
map10 = mapk(user_sequences, all_recommendations, k=10)
ndcg10 = mean_ndcg(user_sequences, all_recommendations, k=10)

print(f"MAP@10: {map10:.4f}")
print(f"NDCG@10: {ndcg10:.4f}")
```
