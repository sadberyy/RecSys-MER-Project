```python
import pandas as pd
import re
```

```python
feature_df = pd.read_csv("opensmile_functionals.csv")
musicinfo_df = pd.read_csv("Music Info.csv")
```

```python
feature_df["track_id"] = feature_df["file"].apply(lambda x: re.split(r"[\.\-]+", x)[-2])
```

```python
feature_df
```

```python
feature_df = feature_df.join(
    musicinfo_df[["track_id", "energy", "valence"]].set_index("track_id"), on="track_id"
)
```

```python
feature_df
```

```python
def classify_emotions(energy, valence):
    e = energy - 0.5
    v = valence - 0.5
    if v > 0 and e > 0:
        return "Joy"
    elif v < 0 and e > 0:
        return "Anger"
    elif v < 0 and e < 0:
        return "Sad"
    else:
        return "Pleasure"
```

```python
feature_df["emotion"] = feature_df.apply(
    lambda x: classify_emotions(x.energy, x.valence), axis=1
)
```

```python
feature_df
```

```python
X = feature_df.drop(
    ["file", "start", "end", "track_id", "energy", "valence", "emotion"], axis=1
)
```

```python
y = feature_df["emotion"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
!pip install catboost
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
from sklearn.utils.class_weight import compute_class_weight

weights = compute_class_weight(
    class_weight="balanced", classes=y_train.unique(), y=y_train
)
```

```python
cb_model = CatBoostClassifier(
    random_state=42,
    loss_function="MultiClassOneVsAll",
    task_type="GPU",
    eval_metric="TotalF1:average=Weighted",
)
```

```python
eval_set = Pool(X_test, y_test)
```

```python
cb_model.fit(X_train, y_train, eval_set=eval_set)
```

```python
train_losses = pd.read_csv("./catboost_info/learn_error.tsv", sep="\t")
test_f1s = pd.read_csv("./catboost_info/test_error.tsv", sep="\t")
```

```python
experiment.log_curve(
    "train_loss_vs_iteration",
    x=train_losses["iter"],
    y=train_losses["MultiClassOneVsAll"],
)
experiment.log_curve(
    "test_f1_vs_iteration", x=test_f1s["iter"], y=test_f1s["TotalF1:average=Weighted"]
)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, cb_model.predict(X_test)))
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
oversample = RandomOverSampler()

X_train_oversample, y_train_oversample = oversample.fit_resample(X_train, y_train)
```

```python
y_train_oversample.value_counts()
```

```python
cb_model_oversampled = CatBoostClassifier(
    random_state=42, task_type="GPU", eval_metric="TotalF1:average=Weighted"
)
```

```python
cb_model_oversampled.fit(X_train_oversample, y_train_oversample, eval_set=eval_set)
```

```python
train_losses = pd.read_csv("./catboost_info/learn_error.tsv", sep="\t")
test_f1s = pd.read_csv("./catboost_info/test_error.tsv", sep="\t")
```

```python
experiment.log_curve(
    "train_loss_vs_iteration_oversampled",
    x=train_losses["iter"],
    y=train_losses["MultiClass"],
)
experiment.log_curve(
    "test_f1_vs_iteration_oversampled",
    x=test_f1s["iter"],
    y=test_f1s["TotalF1:average=Weighted"],
)
```

```python
print(classification_report(y_test, cb_model_oversampled.predict(X_test)))
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
X_train_undersample, y_train_undersample = RandomUnderSampler().fit_resample(
    X_train, y_train
)
```

```python
cb_model_undersampled = CatBoostClassifier(
    random_state=42, task_type="GPU", eval_metric="TotalF1:average=Weighted"
)
cb_model_undersampled.fit(X_train_undersample, y_train_undersample, eval_set=eval_set)
```

```python
train_losses = pd.read_csv("./catboost_info/learn_error.tsv", sep="\t")
test_f1s = pd.read_csv("./catboost_info/test_error.tsv", sep="\t")
```

```python
train_losses
```

```python
experiment.log_curve(
    "train_loss_vs_iteration_undersampled",
    x=train_losses["iter"],
    y=train_losses["MultiClass"],
)
experiment.log_curve(
    "test_f1_vs_iteration_undersampled",
    x=test_f1s["iter"],
    y=test_f1s["TotalF1:average=Weighted"],
)
```
