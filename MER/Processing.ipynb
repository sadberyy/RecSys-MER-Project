```python
file = "CSVs/Blues/Blues-TRACOHF128F1498509.csv"
import pandas as pd
import numpy as np
columns_csvs = pd.read_csv(file).drop(["file", "start", "end"], axis = 1).columns
```

```python
columns = list(columns_csvs)
columns.insert(0, "track_id")
columns.append("genre")
columns
```

```python
df = pd.DataFrame(columns = columns)
```

```python
import re
for row, path in enumerate(Path('CSVs').rglob("*.csv")):
    track_id = re.split("\.|-", path.name)[1]
    genre = re.split("\.|-", path.name)[0]
    arr = list()
    arr.append(track_id)
    arr+= list(pd.read_csv(path)[columns_csvs].mean())
    arr.append(genre)
    df.loc[len(df.index)] = arr
    
```

```python
df
```

```python
annot = pd.read_csv("Music Info.csv")[["track_id", "name","tags", "artist", "energy", "valence"]]
```

```python
annot
```

```python
df_annot = df.join(annot.set_index("track_id"), on = "track_id")
```

```python
df_annot.to_csv("opensmile_features.csv")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 
```

```python
X = df_annot.drop(["track_id", "name", "artist", "energy", "valence", "genre", "tags"], axis = 1)
y = df_annot["genre"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
clf = CatBoostClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
print(classification_report(y_test, clf.predict(X_test)))
```

```python
clf.save_model("genre_classification.cbm")
```

```python
import seaborn as sns
sns.scatterplot(data = df_annot, x= df_annot["valence"], y = df_annot["energy"])
```

```python
from catboost import CatBoostRegressor
reg_energy = CatBoostRegressor(random_state = 42)
reg_valence = CatBoostRegressor(random_state = 42)
energy = df_annot["energy"]
valence = df_annot["valence"]
X_train, X_test, energy_train, energy_test, valence_train, valence_test = train_test_split(X, energy, valence, random_state=42, test_size=0.2)
reg_energy.fit(X_train, energy_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(energy_test, reg_energy.predict(X_test)))
reg_energy.save_model("energy_regression.cbm")
```

```python
reg_valence.fit(X_train, valence_train)
```

```python
print(r2_score(valence_test, reg_valence.predict(X_test)))
reg_valence.save_model("valence_regression.cbm")
```

```python
def classifify_emotions(energy, valence):
    e = energy - 0.5
    v = valence - 0.5
    if v > 0 and e > 0:
        return "Joy"
    elif v < 0 and e > 0:
        return "Anger"
    elif v < 0 and e < 0:
        return "Sad"
    else:
        return "Pleasure"
```

```python
annot["emotion"] = annot.apply(lambda x: classifify_emotions(x.energy, x.valence), axis = 1)
```

```python
annot.to_csv("music_arousal_valence.csv")
```
