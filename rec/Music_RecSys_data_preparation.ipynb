```python
import pandas as pd

data = pd.read_csv("Music Info.csv")
data
```

```python
data.isnull().sum() #пропуски
```

```python
data.info()
```

```python
genres_list = ['classical', 'ambient', 'electronic', 'folk', 'rap', 'indie', 'pop', 'instrumental']

def extract_genre(tag): #функция для заполнения жанров по тегам
    if isinstance(tag, str):
        if 'rock' in tag:
            return 'rock'
        elif 'metal' in tag:
            return 'metal'
        for genre in genres_list:
            if genre in tag:
                return genre
    return ''

data['genre'] = data['tags'].apply(extract_genre)
```

```python
data['genre'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier #заполнение пропусков

train = data[data['genre'] != '']
test = data[data['genre'] == '']

X_train = train[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
X_test = test[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
y_train = train['genre']

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)

model = RandomForestClassifier()
model.fit(X_train, y_train)

data.loc[data['genre'] == '', 'genre'] = model.predict(X_test)
```

```python
data['genre'].value_counts()
```

```python
from sklearn.metrics import accuracy_score, classification_report #проверка метрик на сущетсвующих до этого значениях жанра

train_sample = train.sample(frac=0.8, random_state = 42)
test_sample = train.drop(train_sample.index)

X_train = train_sample[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
X_test = test_sample[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]

X_train = pd.get_dummies(X_train)
y_train = train_sample['genre']
X_test = pd.get_dummies(X_test)
y_test = test_sample['genre']

X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print('accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Распределение длительности треков

```python
plt.boxplot(data['duration_ms']/60000)
plt.ylabel('Количество')
plt.title('Анализ выбросов в длительности треков')
plt.show()
```

В рамках разумного, нет отрицательных значений, нет очень больших, треки реально по 30+ минут идут

```python
Q1 = data['duration_ms'].quantile(0.25)
Q3 = data['duration_ms'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['duration_ms'] < lower_bound) | (data['duration_ms'] > upper_bound)]
print(f"Количество выбросов: {len(outliers)}")
print(f"Количество выбросов в процентах: {len(outliers)/len(data) * 100}")
```

Создаем фейковый timestamp для последовательности взаимодействия юзера с песней, это нужно для SASRec & BERT4Rec

```python
interactions = data
interactions
```

```python
import numpy as np

start_date = pd.Timestamp("2025-01-01")
end_date = pd.Timestamp("2025-01-31")

interactions['timestamp'] = np.random.randint(start_date.timestamp(), end_date.timestamp(), size=len(interactions))

interactions['timestamp'] = pd.to_datetime(interactions['timestamp'], unit="s")
```

```python
interactions = interactions[['user_id', 'track_id', 'timestamp']]
```

```python
uid_encoder = {uid: i + 1 for i, uid in enumerate(set(interactions['user_id']))}
iid_encoder = {iid: i + 1 for i, iid in enumerate(set(interactions['track_id']))}

interactions['user_id'] = interactions['user_id'].map(uid_encoder)
interactions['track_id'] = interactions['track_id'].map(iid_encoder)
```

```python
interactions = interactions.rename(columns={'track_id': 'item_id'})
interactions.head()
```
